1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.
- Relative cell reference
   - It changes automatically when copied to another cell.
   - It is used for repeacting calculations for rows and columns.

- Absolute cell reference
   - It always remains fixed no matter where the formula is copied.
   - It prevents errors when we always want the same value.
   - e.g. - tax rates, constant values and so on.

- Mixed cell reference
   - It is a combined form of relative and absolute cell referencing
   - helpful in tables and multiplying both rows and columns.
   - It gives flexibility in row and column operations.
